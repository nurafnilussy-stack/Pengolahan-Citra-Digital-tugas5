```python
# ===============================================
# LOW-PASS FILTER (LPF) — FINAL NO ERROR VERSION
# ===============================================

from google.colab import files
import cv2
import numpy as np
from PIL import Image
from matplotlib import pyplot as plt

# 1. UPLOAD FILE
print("Silakan upload gambar...")
uploaded = files.upload()

# Ambil nama file yang diupload
filename = list(uploaded.keys())[0]
print("File yang diupload:", filename)

# 2. KONVERSI WEBP → JPG (agar OpenCV bisa membaca)
if filename.lower().endswith(".webp"):
    img_webp = Image.open(filename)
    img_webp = img_webp.convert("RGB")
    img_webp.save("image.jpg")
    img_path = "image.jpg"
else:
    img_path = filename

# 3. LOAD GAMBAR
img = cv2.imread(img_path, 0)

if img is None:
    raise ValueError("Gambar gagal dibaca! Pastikan format benar.")

# 4. LOW-PASS FILTER (LPF)
kernel_lpf = np.ones((3,3), np.float32) / 9
lpf_result = cv2.filter2D(img, -1, kernel_lpf)

# 5. TAMPILKAN HASIL
plt.figure(figsize=(10,5))
plt.subplot(1,2,1); plt.imshow(img, cmap='gray'); plt.title("Original")
plt.subplot(1,2,2); plt.imshow(lpf_result, cmap='gray'); plt.title("Low-Pass Filter (LPF)")
plt.show()
```

```python
# ===============================================
# HIGH-PASS FILTER (HPF) — FINAL NO ERROR VERSION
# ===============================================

from google.colab import files
import cv2
import numpy as np
from PIL import Image
from matplotlib import pyplot as plt

# 1. UPLOAD FILE
print("Silakan upload gambar...")
uploaded = files.upload()

# Ambil nama file upload
filename = list(uploaded.keys())[0]
print("File yang diupload:", filename)

# 2. KONVERSI WEBP → JPG supaya OpenCV tidak error
if filename.lower().endswith(".webp"):
    img_webp = Image.open(filename)
    img_webp = img_webp.convert("RGB")
    img_webp.save("image.jpg")
    img_path = "image.jpg"
else:
    img_path = filename

# 3. LOAD GAMBAR
img = cv2.imread(img_path, 0)

if img is None:
    raise ValueError("Gambar gagal dibaca! Cek format file.")

# 4. HIGH-PASS FILTER (HPF)
kernel_hpf = np.array([
    [-1, -1, -1],
    [-1,  8, -1],
    [-1, -1, -1]
])

hpf_result = cv2.filter2D(img, -1, kernel_hpf)

# 5. TAMPILKAN HASIL
plt.figure(figsize=(10,5))
plt.subplot(1,2,1); plt.imshow(img, cmap='gray'); plt.title("Original")
plt.subplot(1,2,2); plt.imshow(hpf_result, cmap='gray'); plt.title("High-Pass Filter (HPF)")
plt.show()
```

```python
# ===============================================
# BAND-PASS FILTER (BPF) — FINAL NO ERROR VERSION
# ===============================================

from google.colab import files
import cv2
import numpy as np
from PIL import Image
from matplotlib import pyplot as plt

# 1. UPLOAD FILE GAMBAR
print("Silakan upload gambar...")
uploaded = files.upload()

# Nama file
filename = list(uploaded.keys())[0]
print("File yang diupload:", filename)

# 2. KONVERSI WEBP → JPG AGAR OPENCV BISA MEMBACA
if filename.lower().endswith(".webp"):
    img_webp = Image.open(filename)
    img_webp = img_webp.convert("RGB")
    img_webp.save("image.jpg")
    img_path = "image.jpg"
else:
    img_path = filename

# 3. LOAD GAMBAR
img = cv2.imread(img_path, 0)

if img is None:
    raise ValueError("Gambar gagal dibaca! Cek ekstensi file.")

# 4. BAND-PASS FILTER (BPF)
kernel_bpf = np.array([
    [-1, -1, -1],
    [-1,  9, -1],
    [-1, -1, -1]
])

bpf_result = cv2.filter2D(img, -1, kernel_bpf)

# 5. TAMPILKAN HASIL
plt.figure(figsize=(10,5))
plt.subplot(1,2,1); plt.imshow(img, cmap='gray'); plt.title("Original")
plt.subplot(1,2,2); plt.imshow(bpf_result, cmap='gray'); plt.title("Band-Pass Filter (BPF) — Sharpening")
plt.show()
```
